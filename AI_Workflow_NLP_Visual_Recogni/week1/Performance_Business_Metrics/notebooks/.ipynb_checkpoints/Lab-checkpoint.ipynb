```python
import random

def f(x):
    df_id = x.copy(deep=True)
    df_id['comments'] = np.nan
    df_id['sentiments'] = np.nan
    customer_id = x["customer_id"].iloc[0]
    nb_comment = nb_comments[customer_id]
    future_churned = x["subscription_stopped"].max()
    for i in range(nb_comment) : 
        if future_churned == 1 :
            if random.random() < 0.9 :
                # choose negative comment
                comment = corpus_neg.pop(random.randint(0,len(corpus_neg)-1))
                sentiment = 0
                idx = (len(x)-1) - min(int(np.random.exponential(scale=len(x)/7, size=None)), len(x)-1)
            else :
                # choose positive comment
                comment = corpus_pos.pop(random.randint(0,len(corpus_pos)-1))
                sentiment = 1
                idx = random.randint(0, len(x)-1)
        else :
            if random.random() < 0.9 :
                # choose positive comment
                comment = corpus_pos.pop(random.randint(0,len(corpus_pos)-1))
                sentiment = 1
                idx = random.randint(0, len(x)-1)
            else :
                # choose negative comment
                comment = corpus_neg.pop(random.randint(0,len(corpus_neg)-1))
                sentiment = 0
                idx = random.randint(0, len(x)-1)
                
        
        df_id.iloc[idx, 5] = comment
        df_id.iloc[idx, 6] = sentiment
    return df_id

df = stream.groupby('customer_id').apply(f).reset_index(drop=True).copy(deep=True)
```

#### preprocessing  

```python
import spacy
STOPLIST = text.ENGLISH_STOP_WORDS
STOPLIST = set(list(STOPLIST) + ["foo"])

def lemmatize_document(doc, stop_words=None):
    """
    takes a list of strings where each string is a document
    returns a processed list of strings
    """
    
    if not stop_words:
        stop_words = set([])
  
    ## ensure working with string
    doc = str(doc)

    # First remove punctuation form string
    if sys.version_info.major == 3:
        PUNCT_DICT = {ord(punc): None for punc in punctuation}
        doc = doc.translate(PUNCT_DICT)

    # remove unicode
    clean_doc = "".join([char for char in doc if char in printable])
            
    # Run the doc through spaCy
    doc = nlp(clean_doc)

    # Lemmatize and lower text
    tokens = [re.sub("\W+","",token.lemma_.lower()) for token in doc ]
    tokens = [t for t in tokens if len(t) > 1]
    
    return ' '.join(w for w in tokens if w not in stop_words)   
    
## example usage
corpus = ['"You can fool some of the people all of the time, and all of the people some of the time, but you can not fool all of the people all of the time". -- Abraham Lincoln']
processed = [lemmatize_document(doc, STOPLIST) for doc in corpus]
print(processed[0])
processed = [lemmatize_document(doc, None) for doc in corpus]
print("\n"+processed[0])
```

#### Sentiment Analysis

```python
def create_sentiment_data(df):
    X = df.dropna()['comments']
    y = df.dropna()['sentiments']
    return X, y
```

```python
X_train, y_train = create_sentiment_data(df_train)
X_test, y_test = create_sentiment_data(df_test)
```

```python
target_names
```

```python
from sklearn.naive_bayes import MultinomialNB

time_start = time.time()
pipe1  = Pipeline(steps=[('tfidf',TfidfVectorizer()),
                         ('nb', MultinomialNB())])

param_grid1 = {
    'tfidf__max_df':[0.4, 0.6, 0.8],
    'tfidf__max_features': [2000, 4000, 6000],
    'nb__alpha':[0.01, 0.1, 1.0, 2.0]   
}

grid1 = GridSearchCV(pipe1, param_grid=param_grid1, cv=5, n_jobs=-1, scoring='accuracy')
grid1.fit(X_train, y_train)
y_pred = grid1.predict(X_test)
print("train time", time.strftime('%H:%M:%S', time.gmtime(time.time()-time_start)))
print(classification_report(y_test, y_pred, target_names=target_names))
print(grid1.best_params_)
```

```python
from sklearn.model_selection import learning_curve
learning_curve(grid1, X_train, y_train)
```

```python
from sklearn.linear_model import SGDClassifier

time_start = time.time()
pipe2  = Pipeline(steps=[('tfidf',TfidfVectorizer()),
                         ('sgd', SGDClassifier(penalty = 'elasticnet'))])

param_grid2 = {
    'tfidf__max_df':[0.4, 0.6, 0.8],
    'tfidf__max_features': [2000, 4000, 6000],
    'sgd__alpha':[0.0001, 0.001, 0.01],
    'sgd__l1_ratio':[ 0.20, 0.40, 0.50, 0.7]
}

grid2 = GridSearchCV(pipe2, param_grid=param_grid2, cv=5, n_jobs=-1)
grid2.fit(X_train, y_train)
y_pred = grid2.predict(X_test)
print("train time", time.strftime('%H:%M:%S', time.gmtime(time.time()-time_start)))
print(classification_report(y_test, y_pred, target_names=target_names))
print(grid2.best_params_)
```

```python
def create_visu(x):
    sent_mean = x[~x['comments'].isna()]['sentiments'].mean()
    sent_pred_mean = x[~x['comments'].isna()]['sentiment_pred'].mean()
    retention = x['subscription_stopped'].mean()*20
    return pd.Series({'sent_mean': sent_mean, 'sent_pred_mean': sent_pred_mean, 'retention': retention})

df_visu_train = df_train.groupby('M-Y').apply(create_visu)
print(df_visu_train.corr())
ax = df_visu_train.plot()
```

#### Simulation : 

In this simulation, we create a simulated stream of comments that has the same characteristics as the stream of comments on the dataset (same proportion of negative/positive comments and same expected number of comments per week). From this simulated data we compute a simulated output of our prediction model based on the accuracy of the model on the test set. We also compute a simulated retention rate based on the statistical link between the churn rate and the average sentiment of the comments.

```python
def pred_churn_rate(x):
    w0 = np.random.normal(0.9248, 0.124)
    w1 = np.random.normal(0.9753, 0.156) 
    return w0 - w1*x 

churn_rate_simu = [pred_churn_rate(x) for x in sent]


count = [int(np.random.normal(np.mean(count), np.std(count))) for i in range (len(sent))]
pos = np.array(sent)*np.array(count)
pos = pos.astype(int) 


recall = 0.94
precision = 0.78

tp_simu = [recall*p for p in pos]

tpfp_simu = [tp/precision for tp in tp_simu]

sent_pred = [x[0]/x[1] + np.random.normal(0, 0.03) for x in zip(tpfp_simu ,count)]


fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(111)

ax.plot(np.arange(len(sent)), sent)
ax.plot(np.arange(len(churn_rate_simu)), churn_rate_simu)
ax.plot(np.arange(len(sent_pred)), sent_pred)
ax.set_ylim((0, 1))
```
